## 1i)
KNN Regression. Best MSE is at K = 11 for the normalized version: RMSE value for k=  11 is: 4.237497850238749
Best MSE is at k = 3 for the raw features: RMSE value for k=  3 is: 4.014979802194319.
I have then plotted the train and test errors from 1-100 resectively. The train error seems to stay low even as K increases in size. 

```python
#KNN using normalization

train , test = train_test_split(df, test_size = 0.3)

x_train = train.drop('PE', axis=1)
y_train = train['PE']

x_test = test.drop('PE', axis = 1)
y_test = test['PE']

#normalizing
scale = MinMaxScaler(feature_range=(0, 1))

x_train_sc = scale.fit_transform(x_train)
x_train = pd.DataFrame(x_train_sc)

x_test_sc = scale.fit_transform(x_test)
x_test = pd.DataFrame(x_test_sc)

mse_test = []
mse_train = []
for K in range(100):
    K = K+1
    model = neighbors.KNeighborsRegressor(n_neighbors = K)

    model.fit(x_train, y_train)  
    pred=model.predict(x_test)
    pred_train=model.predict(x_train)
    error_test = sqrt(mean_squared_error(y_test,pred))
    mse_train.append(np.mean(pred_train != y_train))
    mse_test.append(error_test)
    
#plotting the rmse values against k values
plt.figure(figsize=(15, 9))
plt.plot(range(1,101), mse_test, label = "test_error")
plt.plot(range(1,101), mse_train, label = "train_error")
plt.title('Error Rate K Value')
plt.xlabel('K Value')
plt.ylabel('Mean Error')
plt.legend()
```
